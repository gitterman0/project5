```python
import pandas as pd

# Probeer met puntkomma als scheidingsteken
df = pd.read_csv("Metric Store - ReadyToGo Bikes.csv", encoding="latin1", sep=";")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Readytogo = df[df["Colour Category"] == "Power 7"]
df_Readytogo = df_Readytogo.fillna(0)

```

```python
df = pd.read_excel("Metric Store - InUse Bikes.xlsx")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Inuse = df[df["Colour Category"] == "Power 7"]
df_Inuse = df_Inuse.fillna(0)
```

```python
df = pd.read_excel("Metric Store - Inventory (1).xlsx")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Inventory = df[df["Colour Category"] == "Power 7"]
df_Inventory = df_Inventory.fillna(0)
```

```python
df = pd.read_excel("Metric Store - Broken Bikes (1).xlsx")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Broken = df[df["Colour Category"] == "Power 7"]
df_Broken = df_Broken.fillna(0)
```

```python
df = pd.read_excel("Metric Store - Subscriptions (1).xlsx")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Subscriptions = df[df["Colour Category"] == "Power 7"]
df_Subscriptions = df_Subscriptions.fillna(0)
```

```python
df = pd.read_excel("Metric Store - Missing Bikes.xlsx")
df["Geographical Indicator"] = df["Geographical Indicator"].ffill()
df_Missing = df[df["Colour Category"] == "Power 7"]
df_Missing = df_Missing.fillna(0)
```

```python
import pandas as pd

key_cols = ["Geographical Indicator", "Colour Category"]
month_cols = [c for c in df_Readytogo.columns if c not in key_cols]

df_list = [df_Readytogo, df_Broken, df_Inventory, df_Inuse, df_Subscriptions,df_Missing]
suffixes = ["Readytogo", "Broken", "Inventory", "Inuse","Subscriptions","Missing"]

# Kopieer key_cols naar de nieuwe DataFrame
df_combined = df_Readytogo[key_cols].copy()

# Voeg kolommen per maand toe, interleaved
new_cols = {}
for m in month_cols:
    for df_src, suffix in zip(df_list, suffixes):
        new_cols[f"{m}_{suffix}"] = df_src[m]

# Voeg alle kolommen in één keer toe
df_combined = pd.concat([df_combined, pd.DataFrame(new_cols)], axis=1)

```

```python
df_combined
```

```python
import pandas as pd

# Functie om strings naar float te converteren
def to_float(val):
    if pd.isna(val):
        return 0.0
    val = str(val)
    val = val.replace('.', '')  # verwijder punten als duizendtallen
    val = val.replace(',', '.') # vervang komma door punt
    try:
        return float(val)
    except ValueError:
        return 0.0  # fallback voor rare tekens

# Pas dit toe op alle numerieke kolommen vanaf de derde
for col in df_combined.columns[2:]:
    df_combined[col] = df_combined[col].apply(to_float)

# Bereken totaal per numerieke kolom
num_cols = df_combined.columns[2:]
totaal = pd.DataFrame(df_combined[num_cols].sum()).T
totaal["Geographical Indicator"] = "Totaal"
totaal["Colour Category"] = ""
totaal = totaal[df_combined.columns]

# Voeg totaalregel onderaan
df_combined_with_total = pd.concat([df_combined, totaal], ignore_index=True)

print(df_combined_with_total.tail())
```
